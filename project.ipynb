```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("iris.csv")
df
```

```python
df
```

```python
df = df.drop("Id", axis =1 )
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['Species'] = encoder.fit_transform(df['Species'])
```

```python
df.head()
```

```python
df['Species'].unique()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size=0.2, random_state=5)
```

```python
X_train
```

```python
Y_train
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape 
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class= 'multinomial', solver='lbfgs')
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,accuracy_score

confusion_matrix(Y_test, y_pred)
```

```python
accuracy= accuracy_score(y_pred, Y_test)
accuracy
```

```python
print("Confusion matrix as a dataframe")
pd.DataFrame(confusion_matrix(Y_test, y_pred), columns = list(range(0,3)))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
acc = accuracy_score(Y_test, y_pred)
acc
```

```python
ps = precision_score(Y_test, y_pred, average= None)
ps
```

```python
r = recall_score(Y_test, y_pred, average = "macro")
r
```

```python
f1 = f1_score(Y_test, y_pred, average= "weighted")
f1
```

```python
params ={
    'solver'  : ['lbfgs', 'liblinear', 'saga', 'saga'],
    'penalty' : ['l2', 'l1'],
    'multi_class' :['auto', 'ovr', 'multinomial']
}
```

```python
from sklearn.model_selection import GridSearchCV
model = LogisticRegression()
grid = GridSearchCV(model, param_grid=params, scoring='accuracy', n_jobs=-1, cv = 5 )
```

```python
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
